# MTF in Earth Observation — From Instrument Design to Image Quality

This notebook builds on the generic MTF concepts introduced in Notebook 1.

Here, the same framework is applied to real Earth Observation systems, where:
- the PSF has a known physical origin
- multiple contributors interact and partially mask each other
- measurements are made on imperfect real images

---

**This notebook covers:**

1. **Physical contributors to system MTF**: optics, pixel, jitter, smear, forward linear motion, ground velocity mismatch
2. **The identifiability problem**: why separating contributors from a single MTF curve is ill-conditioned
3. **Slanted edge on real satellite imagery**: Sentinel-2 Band 4
4. **Parametric fit**: fixing known contributors, estimating the residual blur
5. **Binning and aliasing** in EO: Sentinel-2 10m / 20m / 60m
